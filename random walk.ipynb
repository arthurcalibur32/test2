```python
%matplotlib inline
import numpy as np
import random as rd
import matplotlib.pyplot as plt
position=0
walk=[position]
steps=1000
for i in range(steps):
    step=1 if rd.randint(0,1) else -1
    position+=step
    walk.append(position)


plt.plot(walk)
```

```python
w1=[]
w1=walk
steps=1000
for i in range(steps):
    step=1 if rd.randint(0,1) else -1
    position+=step
    w1.append(position)
plt.plot(w1)
```

```python
nsteps=1000
draws=np.random.randint(-10,11,size=nsteps)
#steps=np.where(draws>0,1,-1)
plt.plot(draws.cumsum(),'b')
```
